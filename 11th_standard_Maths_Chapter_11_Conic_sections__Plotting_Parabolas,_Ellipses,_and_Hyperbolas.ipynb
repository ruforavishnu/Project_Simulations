```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
fig, axs = plt.subplots(1, 3, figsize=(18,5))

# Parabola = y^2 = 4ax
a = 1
x = np.linspace(0, 10, 400)
y = np.sqrt(4 * a * x)
axs[0].plot(x, y, label='$y^2 = 4ax$')
axs[0].plot(x, -y, label='$y^2 = -4ax$')
axs[0].set_title('Parabola : $y^2 = 4ax$')
axs[0].grid(True)
axs[0].legend()
axs[0].axhline(0, color='black', lw=0.5)
axs[0].axvline(0, color='black', lw=0.5)


#  Ellipse (x^2/a^2 + y^2/b^2 = 1)
a = 5
b = 2
theta = np.linspace(0, 2*np.pi, 400)
x = a * np.cos(theta)
y = b * np.sin(theta)

axs[1].plot(x,y)
axs[1].set_title('Ellipse: $\\frac{x^2}{25} + \\frac{y^2}{4} = 1$')
axs[1].grid(True)
axs[1].set_aspect('equal')
axs[1].set_xlim(-a-1, a+1)
axs[1].set_ylim(-b-1, b+1)
axs[1].axhline(0, color='black', lw=0.5)
axs[1].axvline(0, color='black', lw=0.5)


# Hyperbola (x^2/a^2 - y^2/b^2 = 1)
a = 3
b = 2
x = np.linspace(-10, -a,  400)
x2 = np.linspace(a+ 0.1, 10, 400)
y1 = b * np.sqrt( (x**2 / a**2) - 1)
y2 = b * np.sqrt( (x2**2 / a**2) - 1)
axs[2].plot(x, y1, 'r')
axs[2].plot(x, -y1, 'r')
axs[2].plot(x2, y2, 'r')
axs[2].plot(x2, -y2, 'r')
axs[2].set_title('Hyperbola: $\\frac{x^2}{9} - \\frac{y^2}{4} = 1$')
axs[2].grid(True)
axs[2].axhline(0, color='black', lw=0.5)
axs[2].axvline(0, color='black', lw=0.5)

plt.tight_layout()
plt.show()
```
