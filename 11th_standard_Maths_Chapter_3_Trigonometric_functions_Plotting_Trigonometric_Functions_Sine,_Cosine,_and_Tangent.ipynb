```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Create the x values (angles in degrees and radians)
x_degrees = np.linspace(-360, 360, 1000) # 1000 points between -360 degree and 360 degree

x_radians = np.radians(x_degrees) # radian conuterpart of x_degrees. Convert degrees to radians for numpy trig functions

```

```python
print('x_degrees:', x_degrees)
print('\n\n')
print('x_radians:', x_radians)
```

```python
# Calc the trig values
sin_x = np.sin(x_radians)
cos_x = np.cos(x_radians)
tan_x = np.tan(x_radians)

```

```python
plt.figure(figsize=(12,6))

plt.plot(x_degrees, sin_x, label='sin(x)', color='blue')

```

```python
plt.plot(x_degrees, cos_x, label='cos(x)', color='red')
```

```python
tan_x[np.abs(tan_x) > 10] = np.nan # Avoid extremely large values by setting them as NaN
plt.plot(x_degrees, tan_x, label='tan(x)',color='green')
```

```python
# Beautify the plotting - everything in one

plt.figure(figsize=(12,6))

plt.plot(x_degrees, sin_x, label='sin(x)', color='blue')

plt.plot(x_degrees, cos_x, label='cos(x)', color='red')

tan_x[np.abs(tan_x) > 10] = np.nan # Avoid extremely large values by setting them as NaN
plt.plot(x_degrees, tan_x, label='tan(x)',color='green')




plt.title('Trigonometric functions: sin(x), cos(x), tan(x)')
plt.xlabel('Angle (degrees)')
plt.ylabel('Function value')
plt.axhline(0, color='black', linewidth=0.5) # Horizontal axis
plt.axvline(0, color='black', linewidth=0.5) # Vertical axis
plt.grid(True)
plt.legend()
plt.show()
```
