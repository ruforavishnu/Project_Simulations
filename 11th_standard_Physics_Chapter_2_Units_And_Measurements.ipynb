```python
import numpy as np
import matplotlib.pyplot as plt
```

 Simulate Measurement Errors

```python
true_value = 100

num_measurements = 1000
errors = np.random.normal(loc=0, scale=2, size=num_measurements)

measured_values = true_value + errors

```

Visualize Error Distribution

```python
plt.figure(figsize=(10,5))
plt.hist(errors, bins=30, color='skyblue', edgecolor='black', alpha=0.7)

plt.xlabel('Measurement Error')
plt.ylabel('Frequency')
plt.title('Distribution of Measurement Errors')

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
