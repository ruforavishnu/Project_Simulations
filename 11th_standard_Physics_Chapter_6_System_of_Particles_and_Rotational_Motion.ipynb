## Chapter 6 - System of Particles and Rotational Motion

### Project - Simulating Projectile Motion (2D)

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
```

```python
g = 9.8
v0 = 20
theta = 45
dt = 0.05
```

```python
theta_rad = np.radians(theta)

v0x = v0 * np.cos(theta_rad)
v0y = v0 * np.sin(theta_rad)

t_flight = (2 * v0y) / g
t = np.arange(0,t_flight, dt)

x = v0x * t
y = v0y * t - 0.5 * g * t**2
```

```python
type(x)
```

```python
fig, ax = plt.subplots(figsize=(7,5))
ax.set_xlim(0, max(x) + 2)
ax.set_ylim(0, max(y) +2)
ax.set_xlabel('Horizontal Distance (m)')
ax.set_ylabel('Vertical Distance (m)')
ax.set_title('Projectile Motion Simulation')

projectile, = ax.plot([], [], 'bo', markersize=8)

def update(frame):
  projectile.set_data([x[frame]], [y[frame]])
  return projectile,

ani = animation.FuncAnimation(fig, update, frames=len(t), interval=50, blit=True)

HTML(ani.to_jshtml())
```
