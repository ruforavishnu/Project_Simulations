```python
import numpy as np
import matplotlib.pyplot as plt

```

Define the Electric Field Function

```python
def electric_field(q, r0, x,y ):
  rx = x - r0[0]
  ry = y - r0[1]

  r_squared = rx**2 + ry**2
  r_squared[r_squared == 0] = 1e-20


  E_x = q * rx / r_squared**1.5
  E_y = q * ry / r_squared**1.5

  return E_x, E_y

```

Create the Grid and Place Charges

```python
x, y = np.meshgrid(np.linspace(-5, 5, 50), np.linspace(-5, 5, 50))
```

```python
charges = [

           (+1, (0,0) ), # single positive charge at origin
           (-1, (1,0))  # second negative charge
]
```

Compute Total Electric Field

```python
Ex, Ey = np.zeros(x.shape), np.zeros(y.shape)

for q, pos in charges:
  ex, ey = electric_field(q, pos, x, y)
  Ex += ex
  Ey += ey

```

Plot the Vector Field

```python
fig , ax = plt.subplots(figsize=(6,6))
ax.streamplot(x, y, Ex, Ey, color=np.log(np.hypot(Ex, Ey)), cmap='inferno', density=1.2)


for q, pos in charges:
  color = 'r' if q > 0 else 'b'
  ax.plot(pos[0], pos[1], 'o', color=color, markersize=15)



ax.set_title('Electric Field lines')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_aspect('equal')
plt.grid(True)
plt.show()
```
