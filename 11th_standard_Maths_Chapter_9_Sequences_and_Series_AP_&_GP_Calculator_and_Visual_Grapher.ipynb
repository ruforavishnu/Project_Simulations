```python
import matplotlib.pyplot as plt
```

```python
def generate_ap(a, d, n):
  ap = [a + i * d for i in range(n)]
  sum_ap = (n / 2) * (2 * a * (n - 1) * d)
  return ap, sum_ap

```

```python
def generate_gp(a, r , n):
  gp = [a * (r**i) for i in range(n)]
  sum_gp = a * (r** n-1 ) / (r - 1) if r != 1 else a * n
  return gp, sum_gp

```

```python
a_ap, d, n = 2, 3, 10
a_gp, r = 1, 2
```

```python
ap, sum_ap = generate_ap(a_ap, d, n)
gp, sum_gp = generate_gp(a_gp, r, n)


```

```python
print(f'AP: {ap}')
print(f'Sum of AP: {sum_ap}')
print(f'\nGP: {gp}')
print(f'Sum of GP: {sum_gp}')

```

```python
plt.figure(figsize=(10,4))
plt.subplot(1, 2, 1)
plt.plot(ap, marker='o')
plt.title('Arithmetic Progression')
plt.xlabel('n')
plt.ylabel('Term value')


plt.subplot(1, 2, 2)
plt.plot(gp, marker='o', color='green')
plt.title('Geometric Progression')
plt.xlabel('n')
plt.ylabel('Term value')

plt.tight_layout()
plt.show()
```
