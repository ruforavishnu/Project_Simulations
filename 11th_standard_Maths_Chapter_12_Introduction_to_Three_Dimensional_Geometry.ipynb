```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
A = np.array([1, 2, 3])
B = np.array([4, 6, 8])
C = np.array([-2, -1, 0])

```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(*A, color='red', label='A(1, 2, 3)')
ax.scatter(*B, color='blue', label='B(4, 6, 8)')
ax.scatter(*C, color='green', label='C(-2, -1, 0)')


ax.plot(  [A[0], B[0]] , [A[1], B[1]] , [A[2], B[2]]  , color='purple', label='Line AB')

distance = np.linalg.norm(B - A)
print(f'Distance AB = {distance:.2f}')


# section point (2:1)
m, n = 2 , 1
section = (n * A + m * B) / (m + n)

ax.scatter(*section, color='orange', label='Point dividing AB 2:1')


ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Points and Line segment')
ax.legend()
plt.show()
```
