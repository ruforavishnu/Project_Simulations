```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
```

```python
# Given parameters
k = 5 # Spring constant (N/m)
F_max = 10 # Maximum force to apply (N)
mass = 1 # Mass attached to spring (kg)

# Time settings
dt = 0.05 # seconds
t_max = 5 # seconds
t = np.arange(0, t_max, dt)
```

Calculate Stretch (Elongation) Over Time

```python
# Force will increase linearly over time
forces = np.linspace(0, F_max, len(t))


# Stretch (x = F/k ) according to Hooke's law
stetches = forces / k

```

Set up the Animation Plot

```python



fig , ax = plt.subplots(figsize=(7,4))
ax.set_xlim(0,12)
ax.set_ylim(-2,2)
ax.set_xlabel('X-axis (m)')
ax.set_yticks([])
ax.set_title("Spring stretching - Hooke's law")


# Draw the wall and mass
wall = plt.Rectangle((0,-0.5), 0.5, 1, color='black')
ax.add_patch(wall)


mass_plot,   = ax.plot([], [], 'ro', markersize=15) # The moving mass
spring_line, =  ax.plot([], [], 'b-', lw=2) # The spring


# Initialization
def init():
  mass_plot.set_data([], [])
  spring_line.set_data([], [])
  return mass_plot, spring_line

# Update function
def update(frame):
  x = 0.5 + stretches[frame] # Position of mass
  mass_plot.set_data([x] , [0] )

  # spring  from wall to mass
  spring_x = np.linspace(0.5 , x, 20)
  spring_y = 0.1 * np.sin(10 * np.pi * (spring_x - 0.5 ))
  spring_line.set_data(spring_x, spring_y)


  return mass_plot, spring_line


ani = animation.FuncAnimation(fig, update, frames=len(t), init_func=init, blit=True, interval = 50)

HTML(ani.to_jshtml())
```
