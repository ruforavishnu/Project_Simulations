```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
# simulation box
BOX_SIZE = 10

#number of particles
N = 20

# Particle properties
positions = np.random.rand(N, 2) * BOX_SIZE
velocities = (np.random.rand(N,2) - 0.5) * 2 # random velocities
```

```python
fig, ax = plt.subplots(figsize=(4,4))
ax.set_xlim(0, BOX_SIZE)
ax.set_ylim(0, BOX_SIZE)
particles, = ax.plot([], [], 'bo', markersize=5)

plt.title('2D Gas molecule simulation')
plt.show()


def update(frame):
  global positions, velocities

  positions += velocities * 0.1

  # handle collisions within walls
  for i in range(N):
    for d in [0,1]: # x and y
      if positions[i, d] <= 0 or positions[i, d] >= BOX_SIZE:
        velocities[i, d] *= -1 # reverse the velocity component when they hit the wall


  #update plot data
  particles.set_data(positions[:, 0], positions[:, 1])
  return particles,


ani = FuncAnimation(fig, update, frames = 200, interval=50, blit=True)

HTML(ani.to_jshtml())
```
