```python
import numpy as np
import matplotlib.pyplot as plt

# 2x + 3y = 6
x = np.linspace(-5, 5, 400)
y = (6 - 2*x) / 3 # solve for y from our eqn

```

```python
plt.plot(x,y, label=r'$2x + 3y = 6$', color='blue')


#Shade the region where 2x + 3y â‰¤ 6
y_fill = np.linspace(-5, 5, 400)
plt.fill_between(x, y_fill, y, where=(y_fill <= y), color='lightblue', alpha=0.5)

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title('Graph of the inequality 2x + 3y <= 6')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
