```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1 # number of moles
R = 8.314 # universal gas constant J / mol . K
T = 300 # temperature in Kelvin
gamma = 1.4 # adiabatic index for air (Cp/Cv)


```

```python
V = np.linspace(0.1, 2, 100)
```

Calculate pressures

```python
P_isothermal = (n * R * T) / V
k =  P_isothermal[0] * (V[0] ** gamma)
P_adiabatic = k / (V ** gamma)
```

Plot the graph

```python
plt.figure(figsize=(10, 6))
plt.plot(V, P_isothermal, label='Isothermal (T=300K)', color='blue')
plt.plot(V, P_adiabatic, label='Adiabatic', color='red', linestyle='--')


plt.title('P-V diagram : Isothermal vs Adiabatic Process')
plt.xlabel('Volume (m)')
plt.ylabel('Pressure (Pa)')
plt.grid(True)
plt.legend()
plt.show()
```
