```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
```

Define constants

```python
rho = 1000 # Density of flui (kg/m3, eg. water)
g = 9.8 # Gravitational acceleration ( m/sÂ²)
h_max = 10 # Maximum depth (m)


# Time settings
dt = 0.05 # seconds
t_max = 5 # seconds
t = np.arange(0, t_max, dt)
```

 Calculate Pressure at Different Depths Over Time

```python
# Simulate an object sinking deeper over time
depths = np.linspace(0, h_max, len(t))

# Pressure at depth (P = rho * g * h)
pressures = rho * g * depths
```

 Set up the Animation Plot

```python
fig, ax = plt.subplots(figsize=(3,4))
ax.set_xlim(-2, 2)
ax.set_ylim(0, 12)
ax.set_xlabel('Horizontal')
ax.set_ylabel('Depth (m)')
ax.invert_yaxis() # Now depth increases downward
ax.set_title('Pressure increases with depth')


# Draw the water surface
ax.axhline(0, color='blue', lw=15)

# Create an object (a circle)
obj, = ax.plot(0, depths[0], 'ro', markersize=15)

# Text to show pressure
pressure_text = ax.text(-1.5, 1, '', fontsize=12, color='black')


# Initialization

def init():
  obj.set_data([], [])
  pressure_text.set_text('')
  return obj, pressure_text

def update(frame):
  obj.set_data([0], [depths[frame]])
  pressure_value = pressures[frame]
  pressure_text.set_text(f'Pressure: {pressure_value:.1f} Pa')
  pressure_text.set_position((-1.5, depths[frame] + 1))
  return obj, pressure_text



ani = animation.FuncAnimation(fig, update, frames=len(t), init_func=init, blit=True, interval=50)

HTML(ani.to_jshtml())


```
