```python
import matplotlib.pyplot as plt
```

```python
def slope(x1, y1, x2, y2):
  return (y2 - y1) / (x2 - x1)

```

```python
def midpoint(x1, y1, x2, y2):
  return ( (x1+x2)/2 , (y1+y2)/2)
```

```python
def distance(x1, y1, x2, y2):
  return ( (x2-x1)**2 + (y2-y1)**2 ) ** 0.5
```

```python
def line_equation(x1, y1, x2, y2):
  m = slope(x1, y1, x2, y2)
  c = y1 - m* x1
  return m, c
```

```python
x1, y1 = 2, 3
x2, y2 = 6, 7
```

```python
m = slope(x1, y1, x2, y2)
mid = midpoint(x1, y1, x2, y2)
dist = distance(x1, y1, x2, y2)

m_eq, c_eq = line_equation(x1, y1, x2, y2)


print(f'Slope: {m}')
print(f'Midpoint: {mid}')
print(f'Distance between A & B: {dist}')
print(f'Line equation: y = {m_eq:.2f}x + {c_eq:.2f}')

```

```python
x_vals = [x1 - 2, x2 + 2]
y_vals = [m_eq * x + c_eq for x in x_vals]

```

```python
plt.figure(figsize=(8,6))
plt.plot(x_vals, y_vals, label='Line AB', color='blue')
plt.scatter([x1, x2], [y1, y2], color='red',label='Points A and B')
plt.scatter(*mid, color='green', label='Midpoint')
plt.text(x1, y1, f'A({x1}, {y1})', fontsize=12)
plt.text(x2, y2, f'B({x2}, {y2})', fontsize=12)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.title('Straight line through two points')
plt.legend()
plt.show()
```
