```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
complex_numbers = [3 + 4j, 1 - 2j,  -2 + 1j,  -3 - 3j,   2 + 0j]
```

```python
# now lets separate real parts and imaginary parts

real_parts = [z.real for z in complex_numbers]
imaginary_parts = [z.imag for z in complex_numbers]
```

```python
# Lets plot it
plt.figure(figsize=(8,8))
plt.axhline(0, color='black') # x axis
plt.axvline(0, color='black') # y axis
plt.grid()


plt.scatter(real_parts, imaginary_parts, color='blue')

# lets label each point
for z in complex_numbers:
  plt.text(z.real + 0.1 , z.imag + 0.1, f"{z}")


plt.title('Complex numbers on the Complex Plane')
plt.xlabel('Real Axis')
plt.ylabel('Imaginary Axis')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.show()

```
