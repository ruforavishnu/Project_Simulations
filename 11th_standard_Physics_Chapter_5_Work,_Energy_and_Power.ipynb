### Project -  Simulating a Block on an Inclined Plane

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
```

```python
m = 2.0
g = 9.8
theta = 30
L = 5.0
dt = 0.05

```

```python
a = g* np.sin(np.radians(theta))

t_max = np.sqrt(2*L / a)
t = np.arange(0, t_max, dt)

```

```python
v = a * t
s = L - 0.5 * a * t**2
```

```python
x = s * np.cos(np.radians(theta))
y = s * np.sin(np.radians(theta))

```

```python
KE = 0.5 * m * v**2
PE = m * g * (L - y)
```

Create the Animation
