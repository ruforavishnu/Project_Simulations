Solve the linear equations:

$$
\begin{cases}
x + y = 7 \\
x - y = 1
\end{cases}
$$

```python
from sympy import symbols, Eq, solve


x,y = symbols('x y')

eq1 = Eq(x + y, 7)
eq2 = Eq(x - y, 1)

solution = solve((eq1, eq2), (x,y))
solution
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_vals = np.linspace(0, 10, 400)

y1 = 7 - x_vals
y2 = x_vals - 1

plt.figure(figsize=(8,6))
plt.plot(x_vals, y1, label='x + y = 7', color='blue')
plt.plot(x_vals, y2, label='x - y = 1', color='green')


plt.scatter(solution[x], solution[y], color='red', zorder=5, label=f'Intersection : ({solution[x]} , {solution[y]} )')

plt.title('Graphical solution of Linear Equations')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='gray', lw=1)
plt.axvline(0, color='gray', lw=1)
plt.grid(True)
plt.legend()
plt.show()
```

### Solve a quadratic equation



$$
\begin{cases}
x^2 - 5x + 6 = 0
\end{cases}
$$

```python
from sympy import symbols, Eq, solve

x = symbols('x')
eq = Eq(x**2 - 5*x + 6, 0)

roots = solve(eq, x)
roots
```

Plot with Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
  return x**2 - 5*x + 6


x_vals = np.linspace(0, 5, 400)
y_vals = f(x_vals)

plt.figure(figsize=(8,6))
plt.plot(x_vals, y_vals, label='y =  x - 5x + 6', color='purple')
plt.axhline(0, color='gray', lw=1)
plt.axvline(0, color='gray', lw=1)


#plot the roots
for r in roots:
  plt.scatter(float(r), 0, color='red', zorder=5)
  plt.text(float(r),0.5, f'Root: {r}',  ha='center', color='red')


# labels
plt.title('Quadratic equation and its roots')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
### Differentiation
```

$$
f(x) = 3x^3 - 5x^2 + 6x - 2
$$


```python
from sympy import symbols, diff

x = symbols('x')
f = 3*x**3 - 5*x**2 + 6*x - 1

#Differentiate f with respect to x
f_prime = diff(f, x)
f_prime
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
  return 3*x**3 - 5*x**2 + 6*x - 2

def f_prime(x):
  return 9*x**2 - 10*x + 6

x_vals = np.linspace(-2, 4, 400)
y_vals = f(x_vals)
y_prime_vals = f_prime(x_vals)


plt.figure(figsize=(10,6))
plt.plot(x_vals, y_vals, label='f(x) = 3x - 5x + 6x - 2', color='blue')
plt.plot(x_vals, y_prime_vals, label="f'(x) = 9x - 10x + 6", color='orange', linestyle='--')

plt.title('Function and its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.axhline(0, color='gray', lw=1)
plt.axvline(0, color='gray', lw=1)
plt.show()

```

### Integrate this

$$
f(x) = 6x^2 -4x + 3
$$

```python
from sympy import symbols, integrate

x = symbols('x')

f = 6*x**2 - 4*x + 3

# compute the indefinite integral
F = integrate(f, x)
F
```

means we have found our solution as follows:


$$
 (6x^2 - 4x + 3) \, dx = 2x^3 - 2x^2 + 3x + C
$$

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad

def f(x):
  return 6*x**2 - 4*x + 3


x_vals = np.linspace(0, 4, 400)
y_vals = f(x_vals)


plt.figure(figsize=(10,6))
plt.plot(x_vals, y_vals, label='f(x) = 6x - 4x + 3', color='green')

# fill the area under the curve
x_fill = np.linspace(0, 3, 300)
plt.fill_between(x_fill, f(x_fill), alpha=0.3, color='green', label='Area under f(x) from 0 to 3')

#compute the definite integral numerically
area, _ = quad(f, 0, 3)

plt.title('Indefinite integral and Area under the curve (Area = {area:.2f})')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.axhline(0, color='gray', lw=1)
plt.axvline(0, color='gray', lw=1)
plt.show()
```
