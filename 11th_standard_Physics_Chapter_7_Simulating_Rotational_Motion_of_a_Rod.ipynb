## Chapter 8: System of Particles and Rotational Motion

###  Simulating Rotational Motion of a Rod

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
```

```python
#Given parameters
L = 5
omega = np.pi / 2 # Angular velocity (rad/sec) -> 90 degrees / second
dt = 0.05 #Time step
t_max = 8 #Total time of simulation (seconds)

#Time array
t = np.arange(0, t_max, dt)

#Angular position as a function of time
theta = omega * t

```

Create the Animation

```python
fig,ax = plt.subplots(figsize=(6,6))
ax.set_xlim(-L-1, L+1)
ax.set_ylim(-L-1, L+1)
ax.set_xlabel('X (m)')
ax.set_ylabel('Y (m)')
ax.set_title('Rotational motion of a Rod')

#Create the rod. A line from (0,0) to (L,0) initially
rod, = ax.plot([0,L], [0,0], 'b-', linewidth=4)


#Animation function
def update(frame):
  #New coordinates after rotation
  x_end = L * np.cos(theta[frame])
  y_end = L * np.sin(theta[frame])
  rod.set_data([0, x_end], [0,y_end])
  return rod,


#Create Animation
ani = animation.FuncAnimation(fig, update, frames=len(t), interval=50, blit=True)


#Display animation
HTML(ani.to_jshtml())
```
