Plot Motion Graphs

```python
plt.figure(figsize=(10,5))


#Displacement Time Graph
plt.subplot(1,2,1)
plt.plot(t, s, color='blue', label='Displacement (s)')
plt.xlabel('Time (s)')
plt.ylabel('Displacement (m)')
plt.title('Displacement vs Time')

plt.grid()
plt.legend()

#Velocity Time Graph
plt.subplot(1,2,2)
plt.plot(t,v, color='red', label='Velocity (v)')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Velocity vs Time')
plt.grid()
plt.legend()


plt.tight_layout()
plt.show()

```

Create Animation


We animate a ball falling by updating its position in each frame.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

# Constants
g = 9.8  # Gravity (m/s²)
u = 0  # Initial velocity (m/s)
t_max = 3  # Max fall time (seconds)
dt = 0.05  # Time step

# Generate time values
t = np.arange(0, t_max, dt)

# Initial height (e.g., 50m)
s = 50 - (u*t + 0.5*g*t**2)
s = np.maximum(0, s)  # Stop at ground level

# Create figure and axis
fig, ax = plt.subplots(figsize=(5, 7))
ax.set_xlim(-1, 1)  # X limits
ax.set_ylim(0, 50)  # Y limits (height)
ax.set_xlabel("X (m)")
ax.set_ylabel("Height (m)")
ax.set_title("Free-Fall Animation")

# Create the falling object (a ball)
ball, = ax.plot(0, s[0], 'ro', markersize=10)

# Animation function
def update(frame):
    ball.set_ydata([s[frame]])  # Update ball position
    return ball,

# Create animation
ani = animation.FuncAnimation(fig, update, frames=len(t), interval=50, blit=True)

# Display animation in Colab
HTML(ani.to_jshtml())  # This ensures animation works inside Google Colab
```
