```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, lambdify

```

```python
x = symbols('x')
f_expr = x**2
f_prime_expr = diff(f_expr, x)
```

```python
f = lambdify(x, f_expr, 'numpy')
f_prime = lambdify(x, f_prime_expr, 'numpy')

```

Lets plot a derivative on a chart


```python
# Choose point a
a = 2
slope = f_prime(a)
b = f(a) - slope * a  # y = mx + b

# create x values
x_vals = np.linspace(a - 3, a + 3, 400)
y_vals = f(x_vals)
tangent_vals = slope * x_vals + b




```

```python
plt.figure(figsize=(8,6))
plt.plot(x_vals, y_vals, label='f(x) =  x²', color='blue' )
plt.plot(x_vals, tangent_vals, label=f'Tangent at x = {a}', linestyle='--', color='red')
plt.scatter([a], [f(a)], color='black', label='Point of Tangency')


plt.title('Function and Tangent Line at x = 2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()


# print derivative at point
print(f'Derivative of f(x) =  x² at x = {a} is {slope}')
```

Thus we have plotted

$f(x) = x^2$

and  the tangent at $\frac{d}{dx}f(x)$

which is  $\frac{d}{dx}(x^2)$
